### Exhibition of analysis from dataset(s) explored, including visual analyses, captions, and useful descripitions


We will be using spacy as the nlp for our dataset to gather the key attributes in order to build a more meaningful summary of all the comments. As the traffic on the usage of e-commerce increases, people will be adding more and more reviews for the products they've purchased which can then be useful for other aspiring consumers.

Instead of skimming through all the reviews, our target audience can quickly understand the reviews via one single review that will be generated from our model.

As part of our approach, we will be following below steps.

1) Initially, we will be categorizing all the words in each of the reviews into three different groups of positive, neutral and negative based on the presence of Adjective-Noun pair where noun will be the feature of the phone and the adjective would be the one provided by the reviewer.

2) Now we will be considering all the gathered Adjective-Noun pair to get the percentage for positive, neutral and negative mentions for the particular feature.

3) Apart from these Adjective-Noun pair, we will also look for other mentions in the reviews for the features of the phones.

4) We will now consider the weightage of all the adjectives to finally describe a feature with one word. Similarly we will be following this approach for all the features identified for the phone from the input JSON dataset.

Below code provides a visualization of the reviews for single product.

```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 

def frequentWordsPlot(x):
    comment_words = ' '
    stopwords = set(STOPWORDS) 

    # iterate through the csv file 
    for val in x: 

        # typecaste each val to string 
        val = str(val) 

        # split the value 
        tokens = val.split() 

        # Converts each token into lowercase 
        for i in range(len(tokens)): 
            tokens[i] = tokens[i].lower() 

        for words in tokens: 
            comment_words = comment_words + words + ' '


    wordcloud = WordCloud(width = 800, height = 800,background_color ='white',stopwords = stopwords,min_font_size = 10).generate(comment_words) 

    # plot the WordCloud image                        
    plt.figure(figsize = (8, 8), facecolor = None) 
    plt.imshow(wordcloud) 
    plt.axis("off") 
    plt.tight_layout(pad = 0) 

    plt.show() 
    
frequentWordsPlot(df.reviews)
```

Below code provides the exploratory data analysis for the count of each word from the reviews.

```python
from nltk import FreqDist
import seaborn as sns
%matplotlib inline

def freq_words(x, terms = 30):
  all_words = ' '.join([text for text in x])
  all_words = all_words.split()
  
  fdist = FreqDist(all_words)
  words_df = pd.DataFrame({'word':list(fdist.keys()), 'count':list(fdist.values())})
  
  # selecting top 20 most frequent words
  d = words_df.nlargest(columns="count", n = terms) 
  plt.figure(figsize=(20,5))
  ax = sns.barplot(data=d, x= "word", y = "count")
  ax.set(ylabel = 'Count')
  plt.show()

freq_words(df['reviews'])
```
